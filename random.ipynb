```python
df_cols = ['year','Name','release', 'rewatch']
lastyear = pd.DataFrame(index=range(0), columns = df_cols)
```

```python
for i in range(len(diary)):
    date = diary.iloc[i,7]
    year = date[0:4]
    if year == '2023':
        new_row = {'year':year, 'Name':diary.iloc[i,1], 'release':diary.iloc[i,2], 'rewatch':diary.iloc[i,5]}
        lastyear = lastyear.append(new_row, ignore_index=True)
```

```python
rewatched = 0
bofa = []
total = 0
#each favourite movie
for i in range(len(faves)):
    faveT = faves.iloc[i,1]
    #check each movie watched in 2022
    for j in range(len(lastyear)):
        #did i watch tha face movie in 2022
        if faveT == lastyear.iloc[j,1]:
            total += 1
            #was it a rewatch?
            if lastyear.iloc[j,3] == 'Yes':
                rewatched += 1
            if faveT not in bofa:
                bofa.append(faveT)
            
            
```

```python
newthisyear = 0
new =[]
#each favourite movie
for i in range(len(faves)):
    faveT = faves.iloc[i,1]
    #check each movie watched in 2022
    for j in range(len(lastyear)):
        #did i watch tha fave movie in 2022
        if faveT == lastyear.iloc[j,1]:
            #was it a rewatch?
            if lastyear.iloc[j,3] != 'Yes':
                if faveT not in new:
                    new.append(faveT)
new
```

```python
Favesthisyear = len(bofa)/len(faves) *100
yearLoved = total/ len(lastyear) *100
rewatches =  rewatched/total * 100
virginal = len(new)/len(faves) *100
print("you loved: {:0.2f}% of the movies you watched this year".format(yearLoved))
print("you've watched {:0.2f}% of your favourite movies list this year(including first experiences and rewatches)".format(Favesthisyear))
#and found: non-rewatched
print("{:0.2f}% of the movies you loved were for the first time".format(100-rewatches))
print("{:0.2f}% of your all time faves were found this year".format(virginal))
```
